```python
from pandas import Series, DataFrame
import operator
import pandas as pd
```

```python
file_dir = 'A1Ratings.csv'
frame = pd.read_csv(file_dir)
frame
```

```python
# Assignment 1-1
frame.mean()
```

```python
# Assignment 1-3
cols = frame.columns
d = dict()
for c in cols:
    d[c] = (frame[c] >= 4).value_counts()[True].astype(float) / frame[c].count() * 100

d = sorted(d.items(), key=operator.itemgetter(1),reverse = True)
print d
```

```python
# Assignment 1-2
cols = frame.columns
temp = frame.fillna(0)
d = dict()
for c in cols:
    d[c] = (frame[c] > 0).value_counts()[True]
    
d = sorted(d.items(), key=operator.itemgetter(1),reverse = True)
print d
```

```python
# Assignment 1-4
cols = frame.columns
temp = frame.fillna(0)
numOfStarwars = (temp['260: Star Wars: Episode IV - A New Hope (1977)'] >0).value_counts()[True]
temp2 = temp[ temp['260: Star Wars: Episode IV - A New Hope (1977)'] >0]
d = dict()
for c in cols:
    d[c] = (temp2[c] > 0).value_counts()[True].astype(float) / numOfStarwars.astype(float) * 100

d = sorted(d.items(), key=operator.itemgetter(1),reverse = True)
print d
```
